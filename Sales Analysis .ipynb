```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
files=[file for file in os.listdir('C:/Users/User/Desktop/Sales_Data/')]
all_months_data=pd.DataFrame()
for i in files:
    df=pd.read_csv("C:/Users/User/Desktop/Sales_Data/" +i)
    all_months_data=pd.concat([all_months_data,df])
all_months_data.to_csv("All_Months_Data_2019.csv",index=False)
```

```python
all_months_data.head()
```

### Cleaning the Data

### Finding NAN rows

```python
nan_df=all_months_data[all_months_data.isna().any(axis=1)]
nan_df.head()
all_months_data=all_months_data.dropna(how='all')
```

### Finding "Or" and delete it

```python
temp_df=all_months_data[all_months_data['Order Date'].str[0:2] != 'Or']
all_months_data=temp_df
all_months_data
```

### ADDING A MONTH COLUMN

```python
all_months_data["Month"]=all_months_data["Order Date"].str[0:2]
all_months_data.head()
```

```python
all_months_data["Month"]=all_months_data["Month"].astype('int32')
all_months_data.head()
```

```python
all_months_data["Quantity Ordered"]=all_months_data["Quantity Ordered"].astype('int32')
```

```python
all_months_data["Price Each"]=all_months_data["Price Each"].astype('float32')
```

### Adding Sales Column

```python
all_months_data["Sales"]=all_months_data["Quantity Ordered"]*all_months_data["Price Each"]
```

```python
all_months_data.head()
```

### Best Sales Month

```python
best_sales_month=all_months_data.groupby('Month').sum()['Sales']
```

```python
month=range(1,13)
plt.bar(month,best_sales_month)
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.show()
```

### Which City has best Sales

### Adding A new column for city

```python
all_months_data["City"]= all_months_data["Purchase Address"].apply(lambda x: x.split(",")[1])
```

```python
all_months_data.head()
```

```python
best_sales_city_df=pd.DataFrame()
best_sales_city_df=all_months_data.groupby('City').sum()
```

```python
cities=[city for city,df in all_months_data.groupby('City')]

plt.bar(cities,best_sales_city_df["Sales"])
plt.xticks(cities,rotation='vertical')
plt.xlabel("Different Cities")
plt.ylabel("Sales in USD")
```

### What time should we display advertisements to maximise likelihood of customer's buying product?

```python
all_months_data["Order Date"]=pd.to_datetime(all_months_data["Order Date"])
```

### Extracting hour from datetime

```python
all_months_data["Hour"]=all_months_data["Order Date"].dt.hour
all_months_data.head()
```

```python
best_hour_df=all_months_data.groupby("Hour").sum()
best_hour_df
```

```python
hours=[hour for hour,df in all_months_data.groupby("Hour")]
plt.bar(hours,best_hour_df['Quantity Ordered'])
```

### What products are sold together more frequently

Products having same order id are most likely to be sold together

```python
df=all_months_data[all_months_data["Order ID"].duplicated(keep=False)]
df["Grouped"]=df.groupby("Order ID")["Product"].transform(lambda x:','.join(x))
df=df[["Order ID","Grouped"]].drop_duplicates()
df.head()
```
